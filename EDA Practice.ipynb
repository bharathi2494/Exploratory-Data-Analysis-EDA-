# Column Transfer and Pipeline

# One Hot Encoder

```python
import pandas as pd

df = pd.DataFrame({
    'age': [25, 32, 47],
    'income': [50000, 64000, 120000],
    'education': ['Bachelor', 'Master', 'PhD'],  # Ordinal
    'city': ['Bangalore', 'Mumbai', 'Delhi'],    # Nominal
})
```

```python
df
```

# step-by-step preprocessing pipeline

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
import sklearn
```

```python
# define column groups
num_features=['age','income']
ord_features=['education']
nominal_features=['city']
```

```python
# define custom order for ordinal encoding
education_order=[['Bachelor','Master','PhD']]
```

```python
#create Transformers
numerical_transformer=StandardScaler()
ordinal_transformer=OrdinalEncoder(categories=education_order)
nominal_transformer=OneHotEncoder(drop=['Mumbai'], handle_unknown='ignore')
```

```python
#combine into columnTransformer
preprocessor=ColumnTransformer(transformers=[
    ('num', numerical_transformer, num_features),
    ('ord', ordinal_transformer, ord_features),
    ('nom', nominal_transformer, nominal_features)
])

# wrap in a pipeline
pipeline=Pipeline(steps=[('preprocessing', preprocessor)])

#Fit and Transform the data
processed=pipeline.fit_transform(df)

                  
                  
```
